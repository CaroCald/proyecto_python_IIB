```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataFile='C:\\Users\\Carolina\\Documents\\GitHub\\proyecto_python_IIB\\proyecto_casas\\jupyter\\data\\items.csv'
```

```python
data=pd.read_csv(dataFile,sep=",")
```

```python
data
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 100
```

```python
data_unidades = data['unidades'].value_counts()
```

```python
arreglo_labels =[]
arreglo_valores=[]
for val, cnt in  data['unidades'].value_counts().iteritems():
    arreglo_labels.append(val)
    arreglo_valores.append(cnt)
```

```python
print(arreglo_labels)
print(arreglo_valores)
```

```python
y_pos = np.arange(len(arreglo_labels[1:10]))
plt.bar(y_pos, arreglo_valores[1:10], align='center', alpha=0.9,  edgecolor='g', color='y')
plt.xticks(y_pos, arreglo_labels[1:10], rotation=45)
plt.ylabel('Numero de unidades')
plt.title('Casas por unidades')
plt.savefig("casas_unidades.pdf")
plt.show()
```
